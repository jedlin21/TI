```python
with open("norm_wiki_sample.txt", "r") as f:
    text = f.read()

withoutSpaces = list(filter( lambda x: len(x)>0, text.split(" ")))
```

```python
def Markov(source, row):
    data = {}
    for index in range(len(source) - row-1):
        sequence = source[index:index+row]
        letter = source[index+row]
        try:
            data[" ".join(sequence)][letter] +=1
        except:
            try:
                data[" ".join(sequence)][letter] = 1
            except:
                data[" ".join(sequence)] = {letter : 1 } 
    
    for key in data.keys():
        sum = np.sum(list(data[key].values()))
        data[key] = {k: v/sum for k, v in data[key].items()}
    return data
```

```python
data = Markov(withoutSpaces, 2)
```

```python
data
```

```python
def generate(seed, data):
    return np.random.choice(list(data[seed].keys()), p=list(data[seed].values()))

def generate2D(seed, data):
    return np.random.choice(list(data[" ".join(seed)].keys()), p=list(data[" ".join(seed)].values()))
```

Wygeneruj przyblienie jzyka angielskiego na podstawie rda Markova
pierwszego rzdu (rda, gdzie prawdopodobiestwo nastpnego symbolu
zaley od 1. poprzedniego).

```python
data1 = Markov(withoutSpaces, 1)
```

```python
seed = np.random.choice(withoutSpaces)
```

```python
seq = [seed]
for _ in range(200):
    seq.append(generate(seq[-1], data1))
print(" ".join(seq))
```

Nastpnie zrb to samo dla rda Markowa drugiego rzdu (tzn. prawdopodobiestwo nastpnego sowa zaley od dwch poprzednich sw).

```python
data2 = Markov(withoutSpaces, 2)
```

```python
seed = np.random.choice(withoutSpaces)
```

```python
seq = [seed]
seq.append(generate(seq[-1], data1))
for _ in range(200):
    seq.append(generate2D(seq[-2:], data2))
print(" ".join(seq))
```
